```python
# Importing required library
import pandas as pd 
# Importing required library
import requests
# Parsing the HTML content using BeautifulSoup
from bs4 import BeautifulSoup
```

```python
# Sending a GET request to the specified URL
web=requests.get("http://www.dodear.com/en/movie?p=1")
```

```python
webs=web.text
```

```python
# Parsing the HTML content using BeautifulSoup
soup=BeautifulSoup(webs,"lxml")
```

```python
soup_pre=soup.prettify
```

```python
# Finding all matching HTML elements
image=soup.find_all("div",{"class":"col-xs-12 list-meta-wrap"})
# Finding the first matching HTML element
len([images.find("a").text for images in image])
```

```python
# Finding all matching HTML elements
complete=soup.find_all("div",{"class":"list-box col-xs-4 col-sm-3 col-md-3 col-lg-2"})
```

```python
movie=[]
rating=[]
post=[]
view=[]

for i in complete:
# Finding the first matching HTML element
    movie.append(i.find("a",class_="list-anc").text)
# Finding the first matching HTML element
    rating.append(i.find("span").text)
# Finding the first matching HTML element
    post.append(i.find("div",class_="list-meta-left").text)
# Finding the first matching HTML element
    view.append(i.find("div",class_="list-meta-right").text)
data={'movies':movie,'ratings':rating,'posts':post,'views':view}
```

```python
df=pd.DataFrame(data)
```

```python
df
```

```python
df["posts"]=df["posts"].str.strip("PD")
```

```python
df["posts"]=df["posts"].str.replace("RD","  ")
```

```python
df["views"]=df["views"].str.strip("Download")
```

```python
df["views"]=df["views"].str.replace("Views","  ")
```

```python
df["views"]=df["views"].str.replace(",","")
```

```python
df_a=df.loc[:, "views"].str.split("  ",expand=True)
```

```python
df_b=df.loc[:, "posts"].str.split("  ",expand=True)
```

```python
df_a.columns=["views","Download"]
```

```python
df_b.columns=["post","Release_Date"]
```

```python
df.drop(["views","posts"],axis=1,inplace=True)
```

```python
df.insert(2,"Posts",df_b["post"])
```

```python
df.insert(3,"Release_Date",df_b["Release_Date"])
```

```python
df.insert(4,"view",df_a["views"])
```

```python
df.insert(5,"Download",df_a["Download"])
```
